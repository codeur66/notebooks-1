## PySTDF

[PySTDF](https://github.com/cmars/pystdf) provides event-based stream parsing of stdf files, along with indexers that can help you rearrange the data into a more useful tabular form, as well as generate missing summary records or new types of derivative records. A stdf file is a stream of records, each record contains several fields holding data. PySTDF parses these records. By means of *Sinks* PySTDF can be easily extended to process stdf data in different ways. *Sinks* are nothing more than stdf record parse event handler classes (see also [PySTDF Tutorial](http://code.google.com/p/pystdf/wiki/Tutorial)). PySTDF ships with several *Sinks*. For this demo we will write a basic *Sink* that will store all the test result data and test limits from a stdf file in a pandas DataFrame. More advanced *Sinks* can for example also store the stdf data in a database, create html pages or a pdf report. If you are new to stdf files and you need to process them, do spend some time on quick-reading the [stdf specification](http://casey.marshall.googlepages.com/stdfv4-spec.pdf). It will help you understand how PySTDF works and how to write your own more advanced *Sink*.

We will now create a very basic *Sink* that will store the stdf data in a pandas DataFrame. The DataFrame will be used for further analysis. Have a look at the [PySTDF Tutorial](http://code.google.com/p/pystdf/wiki/Tutorial) for details on how to write a *Sink*, and the [pandas introduction on a DataFrame](http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dataframe) to learn more on one of the main pandas data structures. The *DataFrameSink* below is the only code that is needed to parse the stdf file into a format of our choice: a pandas DataFrame!

```python
class DataFrameSink(object):
    def __init__(self):
        self.index_tuples = []
        self.series = []
        self.test_nums = []
        self.results = []
        self.limits = {}
        self.data = None
    
    def before_send(self, source, data):
        rectype, fields = data
        if rectype == pir:
            self.on_pir(fields)
        elif rectype == ptr:
            self.on_ptr(fields)
        elif rectype == prr:
            self.on_prr(fields)
    
    def on_pir(self, fields):
        '''Part Information Record'''
        self.test_nums = []
        self.results = []

    def on_ptr(self, fields):
        '''Parametric Test Record'''
        self.test_nums.append(fields[ptr.TEST_NUM])
        self.results.append(fields[ptr.RESULT])
        self.limits[fields[ptr.TEST_NUM]] = [fields[ptr.LO_LIMIT], fields[ptr.HI_LIMIT]]
 
    def on_prr(self, fields):
        '''Part Result Record'''
        if self.results:
            self.series.append(pandas.Series(self.results, self.test_nums))
            self.index_tuples.append((int(fields[prr.PART_ID]), fields[prr.SITE_NUM]))
        
    def after_complete(self, source):
        names = ['part', 'site']
        index = pandas.MultiIndex.from_tuples(self.index_tuples, names=names)
        self.data = pandas.DataFrame(self.series, index=index)
        self.data.columns.name = 'test'
        self.limits = pandas.DataFrame(sink.limits, index=['low', 'high']).T
        self.limits.index.name = 'test'
```

Open the stdf file, create a PySTDF parser, add in instance of our DataFrameSink to it and start parsing. PySTDF needs a while to get the stdf file processed, so if you plan on rerunning the notebook, please be patient.

```python
f = open('demofile.stdf', 'rb')
parser = Parser(inp=f)
sink = DataFrameSink()
parser.addSink(sink)
parser.parse()
f.close()
data = sink.data  # data is a pandas DataFrame holding stdf test data
limits = sink.limits  # limits is also a DataFrame, it holds the test limits
```

We now have a pandas DataFrame object *data* containing the data extracted from the stdf file *demofile.stdf* by the PySTDF parser using our own DataFrameSink. *demofile.stdf* wil no longer be accessed and all data is in memory now and is handled by pandas.

Each column of *data* contains data for a given test number. So there as many columns in *data* as there unique test numbers in the stdf file. Each row of *data* represent data collected for a given part. pandas can handle missing data, so it is not necessary to have data for all tests for each part. 

Note that the index of data not only indicates the part id but also the id at wich site the part was testsed. In our demofile, all parts were tested on site 0. Further on the DataFrame will be modified as if about half of the parts were tested on site1, just for demonstration purposes.

Another DataFrame that is created by our *Sink* is *limits*, it contains the  high and low limits for all the tests.

```python
limits[0:10]
```

We can select some tests and create histograms of the data.

```python
tests = [1000, 1010, 1020, 1030]
data[tests].hist(bins=20);
```

Instead of letting `hist()` handle the xaxis range, we can control it. Zoom in on the range defined by the low and high test limit. To find out the limits of our tests:

```python
limits.ix[tests]
```

```python
range = tuple(limits.ix[tests[0]])
data[tests].hist(bins=20, range=range);
```

Creating boxplots is also possible.

```python
data[tests].boxplot()
ax = plt.gca()
ax.set_ylim(range);
ax.set_xlabel('Testnumber');
```

Compute some statistics.

```python
data[tests].describe()
```

Lets get back to analyzing a single test.

```python
test = 1030
data[test].hist(bins=20, range=range);
plt.gca().set_title('Test %d' % test);
```

Select the right tail values and have a more detaild look. Which parts are in the right tail?

pandas offers [boolean indexing](http://pandas.pydata.org/pandas-docs/stable/indexing.html#boolean-indexing) which is very powerfull to select data based on a given set of conditions.

```python
threshold = -0.65
condition = data[test] > threshold
right_tail = data[test][condition]
pandas.DataFrame(right_tail)
```

Instead of looking at the right tail data in a tabular form, we can for example also make a plot of it. The threshold level is also indicated.

```python
ax = right_tail.plot(style='rd', ylim=(threshold - 0.01, None));
plt.axhline(threshold, linewidth=2, color='g');
```

For demonstration purposes *data* will be modified as if about half of the parts were tested on site 1 and there is a dependency in the test data on which site a sample is tested. For the parts that changed site, tests described above will get a shift of + 0.1.

```python
data_manipulated = data.reset_index()
data_manipulated[400::]['site'] = 1
data_manipulated[400::][tests] += .1
data_manipulated = data_manipulated.set_index(['part', 'site'])
```

This data manipulation is reflected in the histograms of course, double distributions appear.

```python
data_manipulated[tests].hist(bins=20, range=range);
```

pandas offers tools to group the data. Since there is a data shift based on the site a part is tested, grouping by site is an obvious choice. For more complex data sets, grouping by lot, temperature, voltage or whatever test parameter, or a comination of several parameters is very powerfull.

```python
grouped_by_site = data_manipulated[tests].groupby(level=['site'])
```

The grouped data can be used to generate statistics, make histograms and boxplots, etc.

```python
grouped_by_site.describe()
```

```python
frames = []
keys=[]
for key, df in grouped_by_site:
    keys.append(key)
    frames.append(df)
df = pandas.concat(frames, axis=1, keys=keys).T.reorder_levels([1, 0]).sort_index().T
df.boxplot()
plt.gca().set_xlabel('Testnumber | site');
```
